This is a simple code for you to be able to visualize and see your data.

```python
#importing relevant libraries 
import numpy as np
import matplotlib.pyplot as plt
```

```python
num = 50    #random set of data to visualize 
flname=open("ver"+str(num),'r')   #"ver" is the prefix for the data to do with Vertical motion, change this to match your application. 
data = flname.read()
data = data.split('b')    #data saved in format: b'132', meanign 132 in bytes. 

#The following is just a manual piece of code to save the data in numpy arrays
#This is purely based on how your Arduino is printing the data
#Change this bit to match your application. 

data2 = []
for i in data:
    data2.append(i.split(','))
del(data2[0])

for i in range(len(data2)):
    data2[i][-1] = float(data2[i][-1][0:4])
    data2[i][0] = float(data2[i][0][1:5])
    data2[i][1] = float(data2[i][1])
```
